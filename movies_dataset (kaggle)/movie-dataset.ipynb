```python
def multiple_replaces(string, replacements):
    for old, new in replacements:
        string = string.replace(old, new)
    return string
```

```python
cols = multiple_replaces(str(df.columns.values), [('"', ''), ('[', ''), (']', ''), (',', ''), ("'", ''), ('\n', ''), ('Unnamed: 1', ''), ('Unnamed: 2', '')])
cols = cols.strip().split(';')
cols
```

```python
df = df.assign(
    DadosJuntos = lambda f: f.iloc[:, 0] + f.iloc[:, 1]
)
```

```python
df[cols] = df['DadosJuntos'].str.split(';', expand=True)
```
